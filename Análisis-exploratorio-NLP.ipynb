```python
# Importamos librerías

import json
import gzip
import pandas as pd

# Abrimos el dataset base con gzip
# Introducir en path_file la ruta donde hemos descargado el dataset base

path_file = "XXXXX/Video_Games_5.json.gz" 

with gzip.open(path_file, 'rb') as f:
  df = pd.read_json(f, lines=True)

# A partir del dataset base, balanceamos las muestras para reviews positivas (5 y 4) y negativas (1 y 2) con 10.000 muestras para cada puntuación, generando así el dataset de trabajo para esta fase

filtered_df = df[df['overall'].isin([1, 2, 4, 5])]
num_samples = 10000
sampled_df_1 = filtered_df[filtered_df['overall'] == 1].sample(n=num_samples, random_state=1)
sampled_df_2 = filtered_df[filtered_df['overall'] == 2].sample(n=num_samples, random_state=1)
sampled_df_4 = filtered_df[filtered_df['overall'] == 4].sample(n=num_samples, random_state=1)
sampled_df_5 = filtered_df[filtered_df['overall'] == 5].sample(n=num_samples, random_state=1)
result_df = pd.concat([sampled_df_1, sampled_df_2, sampled_df_4, sampled_df_5])

# Guardamos el dataset de trabajo en formato csv:
result_df.to_csv('XXXXX/sampled_data_file.csv', index=False)
```

```python
# Abrimos el dataset de trabajo (introducir la ruta donde lo hemos guardado previamente) y cargamos un dataframe sólo con los campos objeto de análisis ('overall' y 'reviewText'), visualizamos las 5 primeras reviews

result_df = pd.read_csv("XXXXX/sampled_data_file.csv", sep=',')

df_videogames = result_df.loc[:,['overall', 'reviewText']]

df_videogames.head(5)
```

```python
# Comprobamos si hay valores nulos

df_videogames.isnull().sum()

# Eliminamos los valores nulos

df_videogames.dropna(inplace=True)

# Comprobamos, tras las eliminaciones de valores nulos, el balanceo de la base de datos

df_videogames['overall'].value_counts()

# Observamos un ligerísimo desbalanceo que se considera asumible
```

```python
# Agrupamos las reviews positivas (5 y 4 -> 1) y negativas (1 y 2 -> 0) en una nueva variable ('sentiment_label')

sentiment_label = []

for p in df_videogames['overall']:
  if p < 3:
    sentiment_label.append(0)
  else:
    sentiment_label.append(1)

sentiment_label = pd.Series(sentiment_label)
df_videogames['sentiment_label'] = sentiment_label

# Eliminamos la variable 'overall'

df_videogames = df_videogames.drop('overall', axis=1)

# Llegados a este punto consideramos que el dataset de trabajo está listo para su exploración, lo guardamos en formato csv para posteriores usos:
    #df_videogames.to_csv('XXXXX/df_videogames.csv', index=False)
```

```python
# Comenzamos la exploración observando la distribución de reviews por sentimiento

df_videogames['sentiment_label'].value_counts()
```

```python
# Importamos librerías

from collections import Counter

# Observamos la cardinalidad del vocabulario

splitted_reviews = df_videogames['reviewText'].str.lower().str.split()
words = splitted_reviews.apply(pd.Series).stack().reset_index(drop = True)
vocabulary = Counter(words)
len(vocabulary)
```

```python
# Observamos las 20 palabras más frecuentes

vocabulary.most_common(20)

# Se considera que esas 20 palabras más frecuentes no aportan nada al análisis, lo que señala la necesidad de eliminar stopwords
```

```python
# Importamos librerías

!pip install nltk
from nltk import ngrams
from nltk.probability import FreqDist

# Generamos bigramas y trigramas del corpus

bigrams = list(ngrams(words, 2))
trigrams = list(ngrams(words, 3))

# Obtenemos frecuencias de los distintos bigramas y trigramas

bg_freq = FreqDist(bigrams)
tg_freq = FreqDist(trigrams)

# Observamos los 20 bigramas más frecuentes

bg_freq.most_common(20)
```

```python
# Observamos ahora los 20 trigramas más frecuentes

tg_freq.most_common(20)

# La observación de los bigramas y trigramas más frecuentes apunta en la misma dirección que la observación de las palabras más frecuentes: la necesidad de eliminar stopwords
```

```python
# Importamos librerías

!pip install stop-words
from stop_words import get_stop_words

# Comprobamos cómo quedarían las palabras, bigramas y trigramas más frecuentes eliminando las stopwords
# En primer lugar obtenemos las stopwords en inglés

sw_list = get_stop_words('en')

# Eliminamos las stopwords del vocabulario y calculamos de nuevo las palabras más frecuentes

words_cleaned = []

for w in words:
  if w not in sw_list:
    words_cleaned.append(w)
  else:
    pass

vocabulary_cleaned = Counter(words_cleaned)
vocabulary_cleaned.most_common(20)

# Observamos que, eliminando las stopwords, las palabras más frecuentes del corpus son ahora más adecuadas para el análisis. Detectamos, además, que se deben eliminar las palabras derivadas de game
# ('game', 'games', 'game.') por encontrarse entre las más frecuentes y no aportar nada al análisis
```

```python
# Terminamos con las comprobaciones, sin stopwords, para bigramas y trigramas

bigrams_cleaned = list(ngrams(words_cleaned, 2))
trigrams_cleaned = list(ngrams(words_cleaned, 3))

bgcleaned_freq = FreqDist(bigrams_cleaned)
tgcleaned_freq = FreqDist(trigrams_cleaned)

bgcleaned_freq.most_common(20)
```

```python
# Tras los bigramas, comprobamos ahora cómo quedan sin stopwords los trigramas

tgcleaned_freq.most_common(20)

# Concluimos que eliminando stopwords, las palabras, bigramas y trigramas del corpus ganan en capacidad de análisis
```

```python
# Importamos librerías

!pip install wordcloud
import matplotlib.pyplot as plt
from wordcloud import WordCloud

# Para finalizar la exploración generamos un wordcloud con las 50 palabras más comunes del corpus (sin stopwords)

wordcloud = WordCloud(max_font_size=50, max_words=50, background_color="white").generate(' '.join(words_cleaned))
plt.figure(figsize=(12,6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

# Conclusiones

La exploración del dataframe permite concluir que el corpus, para un análisis más eficiente, necesita de una limpieza de stopwords, otros términos frecuentes que no aportan nada al análisis (derivados de 'game') y signos de puntuación que sólo generan "ruido". Sin la observación de palabras, bigramas y trigramas más frecuentes no se podría haber identificado ese "ruido" ni orientado la limpieza del corpus.
